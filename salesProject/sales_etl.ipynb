```python
def fetch_data():
    try:
        engine = create_engine('postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')
        df = pd.read_sql('SELECT * FROM sales_table', engine)
        df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
        return df
    except Exception as e:
        messagebox.showerror("Database Error", f"Failed to fetch data: {e}")
        return None


def plot_monthly_sales():
    df = fetch_data()
    if df is not None:
        monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()

        fig, ax = plt.subplots(figsize=(6, 4))
        monthly_sales.plot(kind='line', marker='o', ax=ax)
        ax.set_title('Monthly Sales Trend')
        ax.set_xlabel('Month')
        ax.set_ylabel('Total Sales')

        canvas = FigureCanvasTkAgg(fig, master=window)
        canvas.draw()
        canvas.get_tk_widget().pack()

def plot_city_monthly_sales():
    df = fetch_data()
    if df is not None:
        selected_city = city_var.get()
        city_monthly_sales = df[df['CITY'] == selected_city].groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()

        fig, ax = plt.subplots(figsize=(6, 4))
        city_monthly_sales.plot(kind='line', marker='o', ax=ax)
        ax.set_title(f'Monthly Sales Trend in {selected_city}')
        ax.set_xlabel('Month')
        ax.set_ylabel('Total Sales')

        canvas = FigureCanvasTkAgg(fig, master=window)
        canvas.draw()
        canvas.get_tk_widget().pack()

def plot_country_monthly_sales():
    df = fetch_data()
    if df is not None:
        selected_country = country_var.get()
        country_monthly_sales = df[df['COUNTRY'] == selected_country].groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()

        fig, ax = plt.subplots(figsize=(6, 4))
        country_monthly_sales.plot(kind='line', marker='o', ax=ax)
        ax.set_title(f'Monthly Sales Trend in {selected_country}')
        ax.set_xlabel('Month')
        ax.set_ylabel('Total Sales')

        canvas = FigureCanvasTkAgg(fig, master=window)
        canvas.draw()
        canvas.get_tk_widget().pack()



window = tk.Tk()
window.title("Sales Data Dashboard")
window.geometry('1280x700+0+0')
window.configure(bg='gray')

# Buttons
btn_plot_sales = tk.Button(window, text="Show Monthly Sales Trend",font=('times new roman',14,'bold'),width=20,cursor='hand2',command=plot_monthly_sales, bg="white")
btn_plot_sales.place(relx=0.05, rely=0.02)

#for city monthly sales
city_var = tk.StringVar()
city_dropdown = tk.OptionMenu(window, city_var, *df['CITY'].unique())
city_dropdown.place(relx=0.47, rely=0.02)

btn_plot_city_sales = tk.Button(window, text="Show City Monthly Sales", command=plot_city_monthly_sales,font=('times new roman',14,'bold'),width=20,cursor='hand2', bg="white")
btn_plot_city_sales.place(relx=0.3, rely=0.02)


#for county monthly sales
country_var = tk.StringVar()
country_dropdown = tk.OptionMenu(window, country_var, *df['COUNTRY'].unique())
country_dropdown.place(relx=0.79, rely=0.02)

btn_plot_country_sales = tk.Button(window, text="Show Country Monthly Sales", command=plot_country_monthly_sales, bg="white",font=('times new roman',14,'bold'),width=22,cursor='hand2')
btn_plot_country_sales.place(relx=0.6, rely=0.02)


# Assuming df is already fetched and ORDERDATE is datetime
city_monthly_sales = df.groupby([df['ORDERDATE'].dt.to_period('M'), 'CITY'])['SALES'].sum()
print(city_monthly_sales.head())

country_monthly_sales = df.groupby([df['ORDERDATE'].dt.to_period('M'), 'COUNTRY'])['SALES'].sum()
print(country_monthly_sales.head())

city_yearly_sales = df.groupby([df['ORDERDATE'].dt.year, 'CITY'])['SALES'].sum()
print(city_yearly_sales.head())

window.mainloop()
```
